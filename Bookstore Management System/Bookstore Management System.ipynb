```python
import random
def readFile(inFile):
    with open(inFile,"r") as inFile:
        inventory=[]
        for line in inFile:
            tempList=line.strip("\n")
            tempList=tempList.split(";")
            for i in tempList:
                inventory.append(i)
    return inventory
def overWrite():
    with open("Sample_db.txt","w") as outFile:
        for i in range(0,len(inventory),6):
            inventory[i]+=";"
            inventory[i+1]+=";"
            inventory[i+2]+=";"
            inventory[i+3]+=";"
            inventory[i+4]+=";"
            inventory[i+5]+="\n"
        for i in range(len(inventory)):
            outFile.write(inventory[i])
def addBook(): # use as: inventory=addbook()
    x=True
    while x:
        ISBN_Number=input("Please enter the ISBN number: ")
        book_Name=input("Please enter the book's title: ")
        author_Name=input("Please enter the author's name: ")
        subjects=input("Please enter the subjects of the book (seperate each subject with a comma): ")
        pub_Date=input("Please enter the publication date: ")
        price=input("Please enter the price of the book followed by the currency sign: ")
        print("Please confirm the following information:\n" +"ISBN: "+ ISBN_Number, "Book name: "+book_Name, "author's name: "+ author_Name, "subjects: "+ subjects, "Publication date: " +pub_Date, "price: "+ price)
        confirmation=input("if the information are correct type Y, if not type N, to quit type Q")
        if confirmation=="Y":
            inventory.append(ISBN_Number)
            inventory.append(book_Name)
            inventory.append(author_Name)
            inventory.append(subjects)
            inventory.append(pub_Date)
            inventory.append(price)
            newBook=input("Do you want to add another book?(Y/N) ")
            if newBook!="Y":
                x=False
        elif confirmation=="Q":
            x= False
    return inventory
def ShowList():
    count=0
    for i in range(0,len(inventory),6):
        count+=1
        print(count,"-", "ISBN: "+ inventory[i], "Book name: "+inventory[i+1], "author's name: "+ inventory[i+2], "subjects: "+ inventory[i+3], "Publication date: " +inventory[i+4], "price: "+ inventory[i+5],"\n")
def remove(book_Number):
    i=(book_Number-1)*6
    for n in range(6):
        inventory.pop(i)
def removeBook():
    ShowList()
    x= True
    while x:
        book_Number=input("Enter the number of the book you wich to remove: Q to quit")
        if book_Number in "qQ":
            x=False
        elif not book_Number.isdigit():
            print("Please enter a valid value")
        else:
            remove(int(book_Number))
            x=False
def buyBook():
    x= True
    ShowList()
    while x:
        book_Number=int(input("Please enter the number of the book you wich to buy. or enter 0 to quit"))
        if book_Number>len(inventory)/6 or not str(book_Number).isdigit():
            print("Please enter a valid number.")
        else:
            if book_Number!=0:
                i=book_Number*6-6
                print("Confirm buying the book with the following details: \n","ISBN: "+ inventory[i], "Book name: "+inventory[i+1], "author's name: "+ inventory[i+2], "subjects: "+ inventory[i+3], "Publication date: " +inventory[i+4], "price: "+ inventory[i+5],"\n")
                confirmation=input("(Y/N)")
                if confirmation=="Y":
                    remove(book_Number)
                    confirmation=input("You are all set, do you want to buy another book? (Y/N) ")
                    if confirmation!="Y":
                        x=False
            else:
                x=False
def findBook():
    x= True
    while x:
        searchMethod=input("please select the searching method: \n to search by ISBN number enter 1 \n to search by the book title enter 2 \n to search by the author's name enter 3 \n to quit enter Q")
        count=0
        if searchMethod=="1":
            searchFor=input("Enter the ISBN number: ")
            for i in range(0,len(inventory),6):
                count+=1
                if inventory[i]==searchFor:
                    print(count,"-", "ISBN: "+ inventory[i], "Book name: "+inventory[i+1], "author's name: "+ inventory[i+2], "subjects: "+ inventory[i+3], "Publication date: " +inventory[i+4], "price: "+ inventory[i+5],"\n")
                    x=False
            if x:
                print("Thers is no book with this ISBN number. Try again.")
        elif searchMethod=="2":
            searchFor=input("Enter the book's name: ")
            for i in range(1,len(inventory),6):
                count+=1
                if searchFor in inventory[i]:
                    print(count,"-", "ISBN: "+ inventory[i-1], "Book name: "+inventory[i], "author's name: "+ inventory[i+1], "subjects: "+ inventory[i+2], "Publication date: " +inventory[i+3], "price: "+ inventory[i+4],"\n")
                    x=False
            if x:
                print("Thers is no book with this author name. Try again.")
        elif searchMethod=="3":
            searchFor=input("Enter the author's name: ")
            for i in range(2,len(inventory),6):
                count+=1
                if  searchFor.upper() == inventory[i].upper():
                    print(count,"-", "ISBN: "+ inventory[i-2], "Book name: "+inventory[i-1], "author's name: "+ inventory[i], "subjects: "+ inventory[i+1], "Publication date: " +inventory[i+2], "price: "+ inventory[i+3],"\n")
                    x=False
            if x:
                print("Thers is no book with this author name. Try again.")
        elif searchMethod=="Q":
            x=False
        else: 
            print("please enter one of the options")
def modifyBook():
    x= True
    while x:
        modifyBook=input("Please enter the number of the book you want to modify, enter S to show the list if books, enter F to search for a specific book, enter Q to quit.")
        if modifyBook in "sS":
            ShowList()
        elif modifyBook in "fF":
            findBook()
        elif modifyBook in "qQ":
            x=False
        elif modifyBook.isdigit():
            modifyBook=int(modifyBook)
            if modifyBook<=((len(inventory)+1)/6):
                i=modifyBook*6-6
                print(modifyBook,"-", "ISBN: "+ inventory[i], "Book name: "+inventory[i+1], "author's name: "+ inventory[i+2], "subjects: "+ inventory[i+3], "Publication date: " +inventory[i+4], "price: "+ inventory[i+5],"\n")
                confirmation=input("Is this the book you want to modify?(Y/N) ")
                if confirmation=="Y":
                    print("to modify the ISBN number enter 1 \n"+"to modify the books name enter 2 \n"+"to modify the authors name enter 3 \n" + "to modify the subjects date enter 4 \n" + "to modify the publication date enter 5 \n" +"to modify the price enter 6 \n"+ "to quit type Q")
                    modifyMethod=input("")
                    if modifyMethod=="1":
                        newValue=input("what is the new ISBN number? ")
                        inventory[i]=newValue
                        x=False
                    elif modifyMethod=="2":
                        newValue=input("what is the new book's name? ")
                        inventory[i+1]=newValue
                        x=False
                    elif modifyMethod=="3":
                        newValue=input("what is the new author's name? ")
                        inventory[i+2]=newValue
                        x=False
                    elif modifyMethod=="4":
                        newValue=input("what is the new subjects? (seperate the subjects with comma) ")
                        inventory[i+3]=newValue
                        x=False
                    elif modifyMethod=="5":
                        newValue=input("what is the new publication date? ")
                        inventory[i+4]=newValue
                        x=False
                    elif modifyMethod=="6":
                        newValue=input("what is the new price? ")
                        inventory[i+5]=newValue
                        x=False
                        
            else:
                print("There is no book with the number", modifyBook)
        else:
            print("Please enter one of the options.")
    return inventory  
def generateRandomBook():
    n=len(inventory)/6
    randomBookNumber=random.randint(1,n+1)
    i=(randomBookNumber-1)*6
    print(randomBookNumber,"-", "ISBN: "+ inventory[i], "Book name: "+inventory[i+1], "author's name: "+ inventory[i+2], "subjects: "+ inventory[i+3], "Publication date: " +inventory[i+4], "price: "+ inventory[i+5],"\n")
def main():
    global inventory
    inventory= "Sample_db.txt"
    inventory=readFile(inventory)
    repeat=True
    while repeat:
        print("Please enter the number to the corresponding service. \n1 show list of the books in the inventory. \n2 search for a book. \n3 buy a book. \n4 add a book. \n5 remove a book. \n6 modify a book. \n7 random book. \n0 Quit.")
        choice=input("The number of the service? ")
        if choice =="1":
            ShowList()
            backToMenu=input("Enter 1 to go back to the menu.\n enter 0 to quit")
            if backToMenu=="0":
                repeat=False
        elif choice == "2":
            findBook()
            backToMenu=input("Enter 1 to go back to the menu.\n enter 0 to quit")
            if backToMenu=="0":
                repeat=False
        elif choice == "3":
            buyBook()
            backToMenu=input("Enter 1 to go back to the menu.\n enter 0 to quit")
            if backToMenu=="0":
                repeat=False
        elif choice == "4":
            inventory=addBook()
            backToMenu=input("Enter 1 to go back to the menu.\n enter 0 to quit")
            if backToMenu=="0":
                repeat=False
        elif choice == "5":
            removeBook()
            backToMenu=input("Enter 1 to go back to the menu.\n enter 0 to quit")
            if backToMenu=="0":
                repeat=False
        elif choice == "6":
            inventory= modifyBook()
            backToMenu=input("Enter 1 to go back to the menu.\n enter 0 to quit")
            if backToMenu=="0":
                repeat=False
        elif choice == "7":
            generateRandomBook()
            backToMenu=input("Enter 1 to go back to the menu.\n enter 0 to quit")
            if backToMenu=="0":
                repeat=False
        elif choice == "0":
            repeat=False
        else:
            print("Please enter a valid numeber. ")
    overWrite()
    print("Thanks for using the system. \nRaed Alshehri 201937930")
    
main()
```
